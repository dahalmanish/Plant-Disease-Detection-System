```python
def image_loader(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    labels = df.iloc[:, -1].values
    images = df.iloc[:,:-1].values

    images = images / 255.0
    return images,labels
    
    
```

```python
def conv_2d(images,kernels,stride=1,padding=0):
    height_im,width_im=images.shape
    height_k,width_k=kernels.shape
    output_height=((height_im-height_k+2*padding)//stride)+1
    output_width=((width_im-width_k+2*padding)//stride)+1
    output_array=np.zeros((output_height,output_width))
    for y in range(0,output_height):
        for x in range(0,output_width):
            output_array[y,x]=np.sum(images[y:y+height_k,x:x+width_k]*kernels)
    return output_array
        
```

```python
def making_2darray():
    images,labels=image_loader(r"C:\Users\AASHIK\Desktop\pyp\backend\train_data1.csv")
    images_height,images_width=images.shape
    for i in range(0,images_height):
        img_array=images[0].reshape(128,128)
        return img_array
```

```python
def main():
    kernel=np.array([[-1,0,1],
           [-2,0,2],
           [-1,0,1]]);
    images,labels=image_loader(r"C:\Users\AASHIK\Desktop\pyp\backend\train_data1.csv")
    images_height,images_width=images.shape
    for i in range(0,images_height):
        img_array=images[i].reshape(128,128)
        output_array=conv_2d(img_array,kernel,stride=1,padding=0)
        #return output_array
        
        relu_array=np.maximum(0,output_array)
        return relu_array
```
